#**What is EDA??**
* EDA stands for Exploratory Data Analysis - it's the step where you explore the data
    * Understand it
    * Discover Pattern
    * Stop Anamolies ---> outlires
    * Generate insight
    * And Decide what to do next

##**EDA Steps:-**


1. Viewing the Data:-
    * head(), tail(), shape(), info()
    * What columns do i have? What type of data?
2. Summary Statistics:-
    * mean, mode, median, std, min, max, quartiles
    * It helps to understand spread and central tendancy
3. Value Counts:-
    * how many unique values in the column?
    * great for categorical columns
4. Missing Value Analysis
    * Where are the gaps?
    * What percent of data is missing?
5. Visualization
    * Histograms --> Distribution of values
    * Boxplots --> Outliers and Spread ---> how much data is spread
    * Bar plots --> Comparisons of categories
    * Correlation heatmap --> Linear relationship between numerical features
    * Scatter Plots --> Bivariate relationships
6. Target Variable Exploration
    * How does your output (like 'changes' in your dataset) relate to the other variables?
    * Like  it will find the relationship btw output and the some inputs


##**EDA Importance**

* You can't clean or preprocess what you don't understand
* It helps you to identify mistakes, biases, or limitations of the data   // mistake mean any null value or all the zero and some are 600 or 700
* It guides the direction of your data cleaning and fetaure engineering.
* It gives you audience or stakeholderss a "Story" or overview of what the data is telling you.


##**Data Cleaning**


1.   Handling missing values:-
      * Check which columns have missing values (nulls)
      * Strategies to handlee:
          * Drop missing row and columns(only if very few)
          * Impute with / fill with:
              * Mean/Median --> for numerical data --> Like age and salary --> Quantitative
              * Mode --> Categorical Data ---> mean like gender where the output only the F or M if the value id null then we fill it with Mode
              * Advances: Linear Regression, KNN and interpolation(for future learning)
2.   Remove Duplicate:-
      * Detect and Drop exect duplicate rows
3.   Fix Data Type:-
      * Convert wrong types (e.g., numbers stored as string and dates as texts)  --> we convert numbers into the int and same as others..
4.   Handle inconsistant categories:-
      * Clean up categorical values like:
          * "Male", "male", "MALE", "M" should all become "male"
          *  "Yes", "yes", "Y" ---> unify to one format
5.  Detect and handle outliers:-
      * Use boxplot, IQR or z-score
      * Handle by :
          * Removing(if clearly wrong)
          * Capping(e.g., to 95th percentile)  ---> if it going more than 95%ile then we remove this

6. Fix logic or domain errors:-
      * E.g., age = -5 is invalid or BMI = 150 likely an error
      * Can replace with mean, median and remove


##**Data Preprocessing**
* To prepare clean data so it can be analyzed or used in the ML model.
* If data cleaning is about cleaning mistake, then Data preprocessing is about transforming valid into a usable format
* Encoding categorical variables convert text labels(like "male", "yes", "southeast") into numbers ---> any of the ML model like linear regression, logistic regression, kn regression they all are not able to read the string we can pass them as a numerical value like male as 0 and female as 1
* We have two common method for this:-
    1. Label encoding (Ordinal):   Good for ordered categories like "Low", "Medium", "High"
    2. One-Hot encoding (Nominal):   For non-ordered categories like region

* Feature transformation (Log, Square root, etc++) Used to handle skewed data, like right skewed and left skewed data.
* Feature Scaling (Normalization or Standardization) Bring numerical values to the same scale ---> especially usefull for distance based algorithm
* Normalization (Min-Max Scaling): Scales values between 0 and 1
* Standardization (Z-score Scaling): Transforms data to have mean 0 and std 1

##**Featuure Engineering**
* Creating new feature and transforming existing ones to expose usefull patterns that ML model can learn from
* Why do we need it? --> Because ML model don't know domain logic -- we have to give them the right signal   ---> like  we using some extra columns, features and all  thing to increase the accuracy
* Common fetaure engineering technique in ML:
    * Mathematical combination
    * Target based flag
    * Binning(when it helps)
    * Time-Based features(if time exists)

##**Feature Selecting (for ML)**
* selecting  the most usefull features and removing the rest.
* Why its important?
    * Reduce noise and overfitting
    * Speeds up training
    * Improve model accuracy
    * Makes model interpretation easier

1. Filter methods(Pure Statics):
    * Correlation matrix --> remove high correlated features
    * Chi-Square test --> categorical vs categorical
    * ANOVA F-test --> numerical vs categorial target
2. Embedded Method (Selecting build into the model)
    * Lasso regression  --> Shrinks coefficients to 0
    * Tree based models (Random Forest, XGBoost)
        * Feature importance scores
